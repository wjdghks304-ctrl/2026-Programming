###2) **코드 실행 방법**
* 코드 셀 옆의 ▶️ 버튼 클릭

* 또는 Shift + Enter 키를 누르면 실행됩니다.

✅ 실행 결과는 바로 코드 셀 아래에 표시됩니다.

### **3) 코드의 실행 순서**
*   하나의 블록에서는 일반 프로그램과 같이 **책 읽는 순서대로** 코드가 실행됩니다.
*   여러 개의 블록은 **실행 버튼을 클릭한 순서대로** 블록이 실행됩니다.
  
  (실행된 블록 상단 왼쪽에는 실행 순서가 숫자로 나타납니다.)

```python
# 코드는 책 읽는 순서대로 실행됩니다.
a = 515+3
b = a-2

print(b)
```

🌟블록은 실행 버튼을 클릭한 **순서대로 실행**됩니다.

1번 블록 -> 2번 블록 순서로 실행해보고,
[런타임] - [런타임 다시 시작] 후에 2번 블록 -> 1번 블록 순서로 실행해 봅시다.

```python
# 1번 블록
c = 5+3
```

```python
# 2번 블록
print(c)
```

##**0-3. 알아두면 좋은 Colab 기능**

- Ctrl + M + B: 코드 셀 아래에 새 셀 추가

- Ctrl + M + A: 코드 셀 위에 새 셀 추가

- Ctrl + Enter: 현재 셀만 실행

- Shift + Enter: 현재 셀 실행 + 다음 셀로 이동

- Alt + Enter: 현재 셀 실행 + 바로 아래에 새 셀 추가
